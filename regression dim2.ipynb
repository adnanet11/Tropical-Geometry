```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- Tropical plane: f(x, y) = max(a + x, b + y, c)
def tropical_plane(x, y, a=0, b=0, c=1):
    return np.maximum.reduce([a + 2*x, b + 3*y, np.full_like(x, c)])

# --- Create grid in 2D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = tropical_plane(X, Y, a=1, b=2, c=0)

# --- Plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.9, edgecolor='k')
ax.set_title("Tropical Plane: f(x, y) = max(1 + x, 2 + y, 0)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- Tropical plane: f(x, y) = max(a + 2x, b + 3y, c)
def tropical_plane(x, y, a=0, b=0, c=1):
    return np.maximum.reduce([a + 2*x, b + 3*y, np.full_like(x, c)])

# --- Create a grid of points
x = np.linspace(-5, 5, 40)
y = np.linspace(-5, 5, 40)
X, Y = np.meshgrid(x, y)
Z = tropical_plane(X, Y, a=1, b=2, c=0)

# --- Flatten for scatter
X_flat = X.ravel()
Y_flat = Y.ravel()
Z_flat = Z.ravel()

# --- Add Gaussian noise to x, y, and z
np.random.seed(0)
noise_x = np.random.normal(0, 0.2, size=X_flat.shape)
noise_y = np.random.normal(0, 0.2, size=Y_flat.shape)
noise_z = np.random.normal(0, 0.2, size=Z_flat.shape)

X_noisy = X_flat + noise_x
Y_noisy = Y_flat + noise_y
Z_noisy = Z_flat + noise_z

# --- Plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the tropical plane (smooth surface)
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.6, edgecolor='k')

# Plot the noisy points
ax.scatter(X_noisy, Y_noisy, Z_noisy, color='red', s=10, label='Noisy Data')

ax.set_title("Noisy Observations from a Tropical Plane")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression

# --- Tropical plane: f(x, y) = max(1 + 2x, 2 + 3y, 0)
def tropical_plane(x, y, a=1, b=2, c=0):
    return np.maximum.reduce([a + 2*x, b + 3*y, np.full_like(x, c)])

# --- Step 1: Generate grid and noisy points
np.random.seed(0)
x = np.linspace(-5, 5, 25)
y = np.linspace(-5, 5, 25)
X, Y = np.meshgrid(x, y)
Z = tropical_plane(X, Y)

# Add noise
noise = np.random.normal(0, 0.3, size=Z.shape)
Z_noisy = Z + noise

# Flatten for clustering
X_flat = X.ravel()
Y_flat = Y.ravel()
Z_flat = Z_noisy.ravel()
points = np.vstack((X_flat, Y_flat, Z_flat)).T

# --- Step 2: Clustering
k = 3  # expected number of affine pieces
kmeans = KMeans(n_clusters=k, random_state=0).fit(points)
labels = kmeans.labels_

# --- Step 3: Fit planes
planes = []
for i in range(k):
    mask = labels == i
    xi = X_flat[mask]
    yi = Y_flat[mask]
    zi = Z_flat[mask]

    reg = LinearRegression().fit(np.vstack([xi, yi]).T, zi)
    a, b = reg.coef_
    c = reg.intercept_
    planes.append((a, b, c))

    print(f"Plane {i+1}: z = {a:.2f}x + {b:.2f}y + {c:.2f}")

# --- Step 4: Plot results
fig = plt.figure(figsize=(12, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot noisy points
ax.scatter(X_flat, Y_flat, Z_flat, color='black', s=5, alpha=0.5, label="Noisy data")

# Plot fitted planes
x_grid = np.linspace(-5, 5, 30)
y_grid = np.linspace(-5, 5, 30)
Xg, Yg = np.meshgrid(x_grid, y_grid)

colors = ['red', 'blue', 'green']
for i, (a, b, c) in enumerate(planes):
    Zg = a * Xg + b * Yg + c
    ax.plot_surface(Xg, Yg, Zg, alpha=0.4, color=colors[i], label=f"Plane {i+1}")

ax.set_title("Tropical Regression: Fitted Planes from Noisy Data")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression

# --- Tropical plane function
def tropical_plane(x, y, a=1, b=2, c=0):
    return np.maximum.reduce([a + 2*x, b + 3*y, np.full_like(x, c)])

# --- Generate grid and noisy data
np.random.seed(0)
x = np.linspace(-5, 5, 25)
y = np.linspace(-5, 5, 25)
X, Y = np.meshgrid(x, y)
Z = tropical_plane(X, Y)

# Add noise
noise = np.random.normal(0, 0.3, size=Z.shape)
Z_noisy = Z + noise

# Flatten for clustering and regression
X_flat = X.ravel()
Y_flat = Y.ravel()
Z_flat = Z_noisy.ravel()
points = np.vstack((X_flat, Y_flat, Z_flat)).T

# --- Clustering
k = 3
kmeans = KMeans(n_clusters=k, random_state=0).fit(points)
labels = kmeans.labels_

# --- Fit planes
planes = []
for i in range(k):
    mask = labels == i
    xi = X_flat[mask]
    yi = Y_flat[mask]
    zi = Z_flat[mask]
    reg = LinearRegression().fit(np.vstack([xi, yi]).T, zi)
    a, b = reg.coef_
    c = reg.intercept_
    planes.append((a, b, c))

# --- Grid for plotting
x_grid = np.linspace(-5, 5, 50)
y_grid = np.linspace(-5, 5, 50)
Xg, Yg = np.meshgrid(x_grid, y_grid)

# Plot 1: each plane separately
fig1 = plt.figure(figsize=(10, 6))
ax1 = fig1.add_subplot(111, projection='3d')
ax1.scatter(X_flat, Y_flat, Z_flat, color='black', s=5, alpha=0.5, label="Noisy data")
colors = ['red', 'blue', 'green']
for i, (a, b, c) in enumerate(planes):
    Zg = a * Xg + b * Yg + c
    ax1.plot_surface(Xg, Yg, Zg, alpha=0.4, color=colors[i])
ax1.set_title("Fitted Planes (Individual)")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("z")

# Plot 2: max-affine surface (tropical reconstruction)
fig2 = plt.figure(figsize=(10, 6))
ax2 = fig2.add_subplot(111, projection='3d')
ax2.scatter(X_flat, Y_flat, Z_flat, color='black', s=5, alpha=0.5, label="Noisy data")

Z_combined = np.maximum.reduce([a * Xg + b * Yg + c for (a, b, c) in planes])
ax2.plot_surface(Xg, Yg, Z_combined, cmap='viridis', alpha=0.7)
ax2.set_title("Tropical Surface (max of all planes)")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_zlabel("z")

plt.tight_layout()
plt.show()
```
